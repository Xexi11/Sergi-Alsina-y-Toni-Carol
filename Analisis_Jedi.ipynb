```python
#iteratiu
def combo(deck):
    deck = str(deck)
    compt = 0
    aux = []
    aux_Max = []
    combo = 0
    for i in deck:
        compt = compt + 1
        if int(i) > int(combo):
            combo = i
            aux.append(i)
        if len(aux) > len(aux_Max):
            aux_Max = aux
            
    return [aux_Max, compt]
    
```

```python
import random
import math
n = [10,100,1000,10000,10000000,10000000000000]
j_ite=[]

steps_loglog=[0,0,0,0,0,0]
steps_logn=[0,0,0,0,0,0]
steps_log = [0,0,0,0,0,0]
for i in range(len(n)):
    j_ite.append(combo(n[i])[1])
    
    
    for j in range(math.ceil(math.log(n[i]))):
             steps_log[i] = steps_log[i] +1
            
    for j in range(math.ceil(math.log(n[i],10))):
             steps_logn[i] = steps_logn[i] +1
    
    for j in range(math.ceil(math.log(math.log(n[i],2),2))+1):
             steps_loglog[i] = steps_loglog[i] +1 
     
```

```python
import matplotlib.pyplot as plt
import numpy as np

plt.plot(n,j_ite, 'g', label='Iteratiu')
plt.plot(n,steps_loglog, 'y', label='loglog((x)')
plt.plot(n,steps_log, 'r', label = 'log(x)')
plt.plot(n,steps_logn, 'b', label = 'logn(10)')
plt.xscale('log')
plt.xlabel('Mida entrada (n)')
plt.ylabel('Operacions')
plt.title('Analisis de la complexitat')
plt.legend()
plt.show()
```
